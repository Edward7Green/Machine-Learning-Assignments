```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('display.max_columns', None) 
pd.set_option('display.max_rows', None)  
pd.set_option('display.max_colwidth', -1)
```

```python
df = pd.read_csv('train.csv')
```

# LET US FIRST START BY EXPLORING THE DATA

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

As we can see there are no missing values in this dataset, we are good to go.

```python
df.corr()
```

We discard the column PurchaseTime because it has an extremely high correlation with the dependant variable (label) that we are trying to predict, additionally we will also discard the column ID because we cannot make a prediction based on a user ID.

```python
df2 = df.drop(['id', 'purchaseTime'], axis=1)
df2.head()
```

```python
# We start by assigning to X all the columns asides from the dependant variable, and to Y only the "label" column
X = df2.loc[:, df2.columns != "label"]
X.head()
Y = df2["label"]
Y.value_counts()
```

We notice that we are dealing with an extremely imbalanced data set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.25, random_state = 322, stratify = Y)
print(len(Y_train))
print(Y_train.value_counts())
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# We use SMOTE in order to fix the imbalanced dataset problem
smote = SMOTE(sampling_strategy = 'minority', random_state = 322)
X_train, Y_train= smote.fit_resample(X_train, Y_train)
```

```python
# Balancing the dataset
print(len(Y_train))
print("There are {} values equals to 1".format(np.count_nonzero(Y_train == 1)))
print("There are {} values equals to -1".format(np.count_nonzero(Y_train == -1)))
```

# MODELLING USING RANDOMFORESTCLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
randomfc = RandomForestClassifier(random_state=322, n_estimators=80)

randomfc.fit(X_train, Y_train)
pred_randomfc = randomfc.predict_proba(X_test)
Y_pred_randomfc = randomfc.predict(X_test)
```

```python
# Printing the Confusion Matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test, Y_pred_randomfc))
```

```python
from sklearn.model_selection import cross_val_score
from numpy import mean
scores = cross_val_score(randomfc, X, Y, scoring='roc_auc')
print('Mean ROC AUC: %.3f' % mean(scores))
```

# MODELLING USING LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
logres = LogisticRegression()
logres.fit(X_train,Y_train)
Y_pred = logres.predict(X_test)

print(logres.score(X_train, Y_train))
print(logres.score(X_test, Y_test))
```

```python
# Printing the Confusion Matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test, Y_pred))
```

```python
from sklearn.model_selection import cross_val_score
from numpy import mean
scores = cross_val_score(logres, X, Y, scoring='roc_auc')
print('Mean ROC AUC: %.3f' % mean(scores))
```

When we compare the two models, we noticed that the prediction accuracy of the Random Forest Classifier is much higher than that of the Logistic Regression. Consequently, we will be using Random Forest on the testing set in order to predict the probability of purchase.
